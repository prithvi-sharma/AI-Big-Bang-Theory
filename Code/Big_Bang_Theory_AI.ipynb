```python
gpt2.download_gpt2(model_name="355M")
```

```python
gpt2.mount_gdrive()
```

```python
file_name = "Big Bang.txt"
```

```python
gpt2.copy_file_from_gdrive(file_name)
```

```python
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              steps=5000,
              restore_from='fresh',
              run_name='run2',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```

```python
gpt2.copy_checkpoint_to_gdrive(run_name='run2')
```

```python
a = gpt2.generate(sess,
              length=500,
              temperature=0.95,
              prefix="Series 21 Episode 1 - The",
              nsamples=1,
              batch_size=1
              )
```
